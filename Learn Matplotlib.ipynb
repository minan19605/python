Add sub plot

```python
for i in range (1,5):
    plt.subplot(2, 2, i)
    plt.text(0.5, 0.5, str((2,2, i)), ha='center', fontsize=10)

#plt.tight_layout()
plt.show()
```

use plt.tight_layout() will make those sub plots have good location between each other.

```python
for i in range (1,5):
    plt.subplot(2, 2, i)
    plt.text(0.5, 0.5, str((2,2, i)), ha='center', fontsize=10)
    plt.grid()

plt.tight_layout()
```

Notes:
1. You must call plt.subplot(2,2,1) to activate all sub plots.
2. plt.grid can add grid for each sub plot. so it should exists in for loop.

You can set parameters: color, linewidth, linestyle to change line styles

```python
x = [1,2,3,4,5,6]
y = [2*i*i for i in x]
plt.plot(x,y, color='green', linewidth=2, linestyle='-');
s = [i*i for i in x]
plt.plot(x,s, color='grey', linestyle=':');
```

you can set parameters: maker, color to change scatter styles.
the marker has value:
'^', '3', 'v', '2'..........

```python
x = [i for i in range(1,10)]
y = [2*i+1 for i in x]
plt.scatter(x,y, color='blue', marker='3')
```

```python
Set transparance of lines. use alpha. range is 0 - 1 . 0 means fully transparance.
```

```python
x = [i for i in range(1,10)]
y1 = [2*i+1 for i in x]
y2 = [3*i for i in x]
y3 = [6*i for i in x]
y4 = [i*i for i in x]
plt.plot(x,y1, alpha = 0.1, label='alpha 0.1')
plt.plot(x, y2, alpha = 0.5, label='alpha 0.5')
plt.plot(x, y3, alpha = 0.9, label='alpha 0.9');
plt.scatter(x, y4, label='legend for scatter')
#plt.legend()
plt.legend(loc='lower right')
```

This example show how to set legend for each figure (3 lines, one scatter)
You can use loc parameter to set location of the legend, including:
upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center.

how to label name in legend
method 1: assume you have 10 items in the legend, you want modify the first itme's label. you can save legend into a var L. then get the label from L and set new value.

```python
x = [i for i in range(1,10)]
y1 = [2*i+1 for i in x]
y2 = [3*i for i in x]
y3 = [6*i for i in x]
y4 = [i*i for i in x]
plt.plot(x,y1, alpha = 0.1, label='alpha 0.1')
plt.plot(x, y2, alpha = 0.5, label='alpha 0.5')
plt.plot(x, y3, alpha = 0.9, label='alpha 0.9');
plt.scatter(x, y4, label='legend for scatter')
L = plt.legend()
L.get_texts()[0].set_text("Hello example")
```

Method 2: define a name list as legend labels name.

```python
x = [i for i in range(1,10)]
y1 = [2*i+1 for i in x]
y2 = [3*i for i in x]
y3 = [6*i for i in x]
y4 = [i*i for i in x]
plt.plot(x,y1, alpha = 0.1, label='alpha 0.1')
plt.plot(x, y2, alpha = 0.5, label='alpha 0.5')
plt.plot(x, y3, alpha = 0.9, label='alpha 0.9');
plt.scatter(x, y4, label='legend for scatter')

#mylabels = ['lucky', 'Love', 'Lucy', 'left', 'long', 'lonely']
mylabels = ['lucky', 'Love']
plt.legend(labels = mylabels) #mylabels will overwrite the existing label
```
